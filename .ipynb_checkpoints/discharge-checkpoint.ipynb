```python


import matplotlib.pyplot as plt
import pandas as pd
from IPython.core.display import HTML
import numpy as np
import scipy as sp

from matplotlib.pyplot import *
%matplotlib inline
 
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python


discharge = pd.read_csv('discharge.csv',sep=',',  parse_dates=[0])
discharge = discharge.set_index('date')
discharge.head()
runoff = discharge.copy()
runoff.columns = ['runoff_mm_day_outlet', 'runoff_mm_day_sfk']
# / cubic m per cubic ft * (s / day) * (km2/m^2) * (mm/m) / (sq km area)
runoff['runoff_mm_day_outlet'] = runoff['runoff_mm_day_outlet'] / 35.3147 * (60*60*24) *  0.000001 * 1000 / 417
runoff['runoff_mm_day_sfk'] = runoff['runoff_mm_day_sfk'] / 35.3147 * (60*60*24) *  0.000001 * 1000 / 642.317
```

$$ runoff \frac{mm}{day} = 
discharge \frac{cubic feet}{second} * 
\frac{1 cubic meter}{35.3147 cubic feet} *
\frac{60*60*24 seconds}{1 day} *
\frac{10^-6 km^2}{1 m^2} *
\frac{1000 mm}{1 m} *
\frac{1}{area of catchment km^2} *
$$

```python

# Create a figure
plt.figure(figsize=(20,5))
plt.plot(discharge.index, discharge.discharge_csf_sfk, label='south fork eel, 248 sq mi drainage, coastal belt')
plt.plot(discharge.index, discharge.discharge_cfs_outlet, label='outlet, 161 sq mi drainage, central belt')
plt.legend(bbox_to_anchor=(.5, 1.1))
plt.ylabel("Mean daily discharge [cfs]")
plt.xlabel("Date")
plt.ylim([1,10000])
#plt.xlim([1,1000])
plt.yscale('log')
#plt.xscale('log')
ax = plt.gca()
```

```python
startDateTime = pd.to_datetime('1990-10-1 00:00:00')
stopDateTime = pd.to_datetime('1991-10-1 00:00:00')
```

```python
discharge = discharge[ (pd.to_datetime(discharge.index) > startDateTime) & (pd.to_datetime(discharge.index) < stopDateTime)]
# Create a figure
plt.figure(figsize=(20,5))
plt.plot(discharge.index, discharge.discharge_csf_sfk, label='south fork eel, 248 sq mi drainage, coastal belt')
plt.plot(discharge.index, discharge.discharge_cfs_outlet, label='outlet, 161 sq mi drainage, central belt')
plt.legend(bbox_to_anchor=(.5, 1.1))
plt.ylabel("Mean daily discharge [cfs]")
plt.xlabel("Date")
plt.ylim([1,10000])
#plt.xlim([1,1000])
plt.yscale('log')
#plt.xscale('log')
ax = plt.gca()
```

```python

# Create a figure
plt.figure(figsize=(20,5))
plt.plot(runoff.index, runoff.runoff_mm_day_sfk, label='south fork eel, 248 sq mi drainage, coastal belt')
plt.plot(runoff.index, runoff.runoff_mm_day_outlet, label='outlet, 161 sq mi drainage, central belt')
plt.legend(bbox_to_anchor=(.5, 1.1))
plt.ylabel("Mean daily runoff [mm/day]")
plt.xlabel("Date")
#plt.ylim([1,10000])
#plt.xlim([1,1000])
plt.yscale('log')
#plt.xscale('log')
ax = plt.gca()
```
